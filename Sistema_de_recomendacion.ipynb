### Sistema de Recomendación.

Importo las librerías

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import fastparquet
```

Cargo el archivo que será usado para hacer el sistema de recomendación.

```python
df_steam_games = pd.read_parquet("output_steam_games_ETL.parquet")
```

```python
df_steam_games
```

Obtengo los datos tipos numéricos.

```python
df_numeric_steam_games = df_steam_games.select_dtypes(include='number')
```

Cargo la matríz de correlación

```python
correlation_matrix = df_numeric_steam_games.corr()
```

Asigno un límite en la correlación y creo la variable donde serán guardadas la columnas a eliminar.

```python
limite = 0.8
drop_set = set()
```

```python
# Bucle donde obtengo las correlaciones mayores a 0.8.
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > limite:
            nombre_columna = correlation_matrix.columns[i]
            drop_set.add(nombre_columna)
            
# Elimino las columnas correlacionadas del df original.
df_steam_reducido = df_steam_games.drop(columns=drop_set)
```

```python
df_steam_reducido # Observo el resultado.
```

```python
df_reducido_numerico = df_steam_reducido.select_dtypes(include='number')
df_reducido_numerico
```

```python
# Obtengo la matríz con los datos numéricos para realizar las comparaciones en el 
# sistema.
similarity_matrix = cosine_similarity(df_reducido_numerico)
similarity_matrix
```

Creo la función del sistema de recomendación con la que obtendré todas las recomendaciones.

```python
def recomendar_juegos(id_game, similarity_matrix, n=5):
    # Encuentra el índice del juego en el DataFrame
    juego_idx = df_steam_reducido[df_steam_reducido["id"] == str(id_game)].index[0]
    print(juego_idx)
    
    # Obtiene las similitudes de ese juego con los demás juegos
    similar_indices = similarity_matrix[juego_idx].argsort()[::-1][1:n+1]
    
    # Retorna los nombres de los juegos recomendados
    return {"Juegos recomendados": df_steam_reducido.iloc[similar_indices][['app_name']]}
```

Compruebo los resultados.

```python
recomendaciones = recomendar_juegos(761140, similarity_matrix)
print(recomendaciones)
```

Exporto los datasets usados enla creación de la función para optimizar su uso.

```python
df_steam_reducido.to_parquet("steam_reducido_para_ml.parquet")
df_reducido_numerico.to_parquet("steam_reducido_numerico_para_ml.parquet")
```

Ahora almaceno todos los valores de los juegos recomendados para ser usados en la API.

```python
# Lista donde serán almacenados todas las recomendaciones.
lista_con_recomendaciones = []

for id_game in df_steam_reducido["id"]:
    
    # Me aseguro de manejar el caso de error donde no exista un id en el dataset.
    if id_game is None:
        continue
    
    # Cantidad de juego recomendados.
    n = 5
    
    # Encuentra el índice del juego en el DataFrame.
    juego_idx = df_steam_reducido[df_steam_reducido["id"] == id_game].index[0]
    
    # Obtiene las similitudes de ese juego con los demás juegos.
    similar_indices = similarity_matrix[juego_idx].argsort()[::-1][1:n+1]
    
    # Guarda los nombres de los juegos recomendados en una lista.
    lista_con_recomendaciones.append([id_game, df_steam_reducido.iloc[similar_indices][['app_name']]])
    
```

```python
# Lista donde serán almacenadas todas las recomendaciones.
lista_de_recomendaciones = []

# Cantidad de juegos recomendados.
n = 5

# Itera sobre los ids de los juegos en el DataFrame reducido.
for id_game in df_steam_reducido["id"]:
    # Verifico si el id del juego es válido.
    if id_game is None:
        continue
    
    # Encuentra el índice del juego en el DataFrame.
    juego_idx = df_steam_reducido[df_steam_reducido["id"] == id_game].index
    
    # Me aseguro de que el índice existe.
    if len(juego_idx) == 0:
        continue
    
    juego_idx = juego_idx[0]
    
    # Obtiene las similitudes de ese juego con los demás juegos.
    similar_indices = similarity_matrix[juego_idx].argsort()[::-1][1:n+1]
    
    # Guardo el id del juego y los nombres de los juegos recomendados en una lista.
    recomendaciones = df_steam_reducido.iloc[similar_indices]["app_name"].tolist()
    lista_de_recomendaciones.append([id_game, recomendaciones])
```

Transformo la lista a un Dataframe para exportar.

```python
df_recomendaciones = pd.DataFrame(lista_de_recomendaciones, columns=["id_game", "recomendaciones"])
```
